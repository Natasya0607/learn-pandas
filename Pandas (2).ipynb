# String IO
string IO simpelnya memudahkan dalam memperlakukan string dalam bentuk file. dari data string tsb, dapat disimpan dalam bentuk csv tanpa perlu kita simpan file fisik terlebih dahulu. string IO bisa menyimpan dalam bentuk file lainnya

```python
from io import StringIO

data = ('col1,col2,col3\n'
        'x,y,1\n'
        'a,b,2\n'
        'c,d,3')
type(data)
```

```python
pd.read_csv(StringIO(data))
```

# Usecols
dalam penggunaan pandas adalah argumen untuk menentukan kolom mana yang ingin dibaca dalam file CSV atau file lainnya.

```python
hi = pd.read_csv(StringIO(data), usecols=['col1', 'col3'])
hi
```

kita bisa menyimpan file dari obj data tersebut ke dalam file csv beneran.

```python
hi.to_csv('dataset.csv')
```

kita dapat menyimpan data dalam bentuk type data object.pandas membaca semua kolom sebagai objek tipe data.

```python
date = pd.read_csv(StringIO(data), dtype = object)
date
```

```python
# hanya memanggil 1 kolom yakni bagian col 1
date['col1']
```

```python
# membuat bentuk obj yang lain
angka = ('a,b,c,d\n'
         '1,2,3,4\n'
         '5,6,7,8\n'
         '9,10,11,12'
)
pd.read_csv(StringIO(angka), dtype = float)
```

dalam membuat dtype kita dapat menegubahnya dalam type float ataupun int. terserah gimana bentuknya (sesuai dengan isinya apakah angka atau string)

```python
hi = pd.read_csv(StringIO(angka), dtype = {'b':int, 'c':float, 'a':'int64'})
hi
```

dalam code di atas kita juga dapat memasukan tipe data dalam jmlh bnyk.

```python
# melihat di dalam tabel di kolom a baris ke 1
hi['a'][1]
```
