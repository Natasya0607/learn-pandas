pada pandas,dapat melakuan pembuatan bingkai tabel yang diambil dari tipe data dictionary. contoh sebagai berikut:

```python
data = {
    'nama depan': ['abc', 'halo'],
    'nama belakang': ['hi', 'baca'],
    'umur': ['21', '22']
}

print(data)
```

```python
#dilanjutkan dengan menggunakan dataframe untuk mengkotakkan menggunakan bingkai data
frame = pd.DataFrame(data) #code disamping digunakan untuk membuat tabel
frame
```

```python
# kita dapat melakukan pembuatan data frame dengan hanya menggunakan satu kolom (kita menanggil hanya satu kolom)
# caranya mirip sekali ketika kita memanggil key pada dictionary.
frame['umur']
```

```python
frame[['umur', 'nama belakang']]
#bahkan kita dapat memanggil beberapa tabel di dalam dataframe tsb dengan menggunakan nama key.
```

```python
#kita dapat membuat atau mencetak bagian kolom. maksudnya di apa apa saja
# nama nama yang ada di colom tersebut.
frame.columns
```

pada pandas terdapat cara untuk mencetak atau mengakses baris berdasarkan numerik atau nomor urut dari kolom disebut dengan *iloc*. sedangkan *loc* digunakan untuk mengakses baris dan kolom dalam DataFrame berdasarkan label baris dan kolom.

```python
frame.iloc[1]
```

```python
frame.iloc[[0,1]]
```

```python
#kita dapat mengambil nilai di salah satu kolom dengan beberapa di stiap baris
#cthnya ketika kita ingin melihat siapa siapa aja namanya didalam tabel tersebut
frame.iloc[[0,1],1]
```

```python
#menggunakan loc
#disini kebetulan bagian kolom tidak diberikan label hanya saja baris
frame.loc[[0, 1], ['umur', 'nama belakang']]
```
